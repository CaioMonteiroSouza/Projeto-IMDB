```python
import pandas as pd

data = pd.read_csv('actors.csv')

data.head(20)
```

```python
df = data
```

```python
resultados = []

# Iterando sobre as linhas do DataFrame
for indice, linha in df.iterrows():
    resultado = linha['Total Gross'] / linha['Number of Movies']
    resultados.append(resultado)

# Encontrando o maior valor e sua posição no DataFrame
maior_valor = max(resultados)
indice_maior_valor = resultados.index(maior_valor)

# Imprimindo os resultados
print("Resultados das divisões:")
for i, resultado in enumerate(resultados):
    print(f"Linha {i+1}: {resultado}")

print(f"\nMaior valor encontrado: {maior_valor} na linha {indice_maior_valor + 1}")
```
